```python
import numpy as np
import pandas as pd
```

```python
deli= pd.read_csv("deliveries.csv")
match= pd.read_csv("matches.csv")
```

```python
match.columns
```

```python
total=deli.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total
```

```python
total=total[total['inning']==1]
```

```python
total#stores the total score of match innings 1
```

```python
merge= match.merge(total,left_on='id',right_on='match_id')
```

```python
merge
```

```python
merge=merge.drop("id",axis=1)
```

```python
#removing teams which are not playing
match['team1'].unique()
```

```python
realteams=['Royal Challengers Bengaluru','Delhi capitals', 'Mumbai Indians', 'Kolkata Knight Riders',
       'Rajasthan Royals', 'Chennai Super Kings', 'Sunrisers Hyderabad','Punjab Kings',
       'Lucknow Super Giants', 'Gujarat Titans',
       ]
```

```python
merge['team1']= merge['team1'].str.replace('Delhi Daredevils','Delhi capitals')
merge['team2']= merge['team2'].str.replace('Delhi Daredevils','Delhi capitals')

merge['team1']= merge['team1'].str.replace('Kings XI Punjab','Punjab Kings')
merge['team2']= merge['team2'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
merge['team1']= merge['team1'].str.replace('Gujarat Lions','Gujarat Titans')
merge['team2']= merge['team2'].str.replace('Gujarat Lions','Gujarat Titans')

merge['team1']= merge['team1'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')
merge['team2']= merge['team2'].str.replace('Royal Challengers Bangalore','Royal Challengers Bengaluru')

merge['team1']= merge['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
merge['team2']= merge['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
merge= merge[merge['team1'].isin(realteams)]
merge= merge[merge['team2'].isin(realteams)]
```

```python
merge
```

```python
match_df=merge[['match_id','city','winner',"total_runs"]]
```

```python
match_df=match_df.merge(deli,left_on="match_id",right_on="match_id")
```

```python
match_df
```

```python
match_df=match_df[match_df["inning"]==2]#filtering second innings
```

```python
match_df['current_score']=match_df.groupby('match_id')['total_runs_y'].cumsum()#cumultv sum in each ball
```

```python
match_df
```

```python
#runs left = total - current
match_df['runs_left']= match_df['total_runs_x']-match_df['current_score']
```

```python
match_df
```

```python
match_df['balls left']=120-(match_df['over']*6+match_df['ball'])
```

```python
match_df
```

```python
#to count no. of wickets after each ball
match_df['player_dismissed']= match_df['player_dismissed'].fillna("0")#null value to 0 string
match_df['player_dismissed'] = match_df['player_dismissed'].apply(lambda x:x if x == "0" else "1")# jauthi player kn name achi dont tch
match_df['player_dismissed'] = match_df['player_dismissed'].astype('int')
wickets =match_df.groupby('match_id')['player_dismissed'].cumsum().values
match_df['wickets'] = 10 - wickets
match_df.head(50)
```

```python
match_df['crr']=match_df['current_score']*6/(120-match_df['balls left'])
```

```python
match_df
```

```python
match_df['rrr']=match_df['runs_left']*6/(match_df['balls left'])
```

```python
match_df
```

```python
def result(row):
    if row['winner']==row['batting_team'] :
        return 1
    else:
        return 0
#jadi batting team and winner same then 1 else 0
        
```

```python
match_df['result']=match_df.apply(result , axis=1)
#apply fun is used to apply the result fun in the dataset
```

```python
match_df
```

```python
final=match_df[['batting_team','bowling_team','city','runs_left','balls left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final
```

```python
#the data has been suffled to reduce biasness 
final=final.sample(final.shape[0])
```

```python
final
```

```python
final.dropna(inplace=True)#dropping the mkissing values
```

```python
final=final[final['balls left']!=0]
```

```python
from sklearn.model_selection import train_test_split
X= final.iloc[:,:-1]
y= final.iloc[:,-1]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2 , random_state=1 )
```

```python
X_train
```

```python
#this will convert strings to numeric
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
#we checked there is no infinite values
final.describe()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',RandomForestClassifier(n_jobs=-1))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```
